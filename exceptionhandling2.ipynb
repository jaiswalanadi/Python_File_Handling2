```python
#4
'''
Lookup Error acts as a base class for the exceptions that occur when a key or index 
used on a mapping or sequence of a list/dictionary is invalid or does not exists.
The two types of exceptions raised are: IndexError. KeyError.
'''
```

```python
#4.1
# Key Error
try:  
    a = {1:'a', 2:'b', 3:'c'}  
    logging.info(a[4])  
except LookupError:  
    logging.error("Key Error Exception Raised.")
else:  
    logging.info("Success, no error!")
```

```python
#4.2
# Index Error
try:  
    a = ['a', 'b', 'c']  
    logging.info(a[4])  
except LookupError:  
    logging.error("Index Error Exception Raised, list index out of range")
else:  
    logging.info("Success, no error!")
```
